## Matplotlib example forkflow

```python
# 0. import matplotlib
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save & show
fig.savefig("plots/sample-plot.png")
```

```python
# Make a plot from dictionary
```

```python
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut Butter Store",
       ylabel="Price ($)");
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), nut_butter_prices.values())
ax.set(title="Horizontal Nut Butter Store",
        xlabel="Price (BGN)")
```

```python
# Make some data for histogram and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

### Two options for subplots

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
a = np.linspace(1, 10, 100)
ax1.plot(a, a ** 2);
ax2.plot(a, np.exp(a))
ax3.scatter(np.random.random(10), np.random.random(10))
ax4.bar(nut_butter_prices.keys(), nut_butter_prices.values());
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

# Plot to each different index
ax[0, 0].plot(a, a ** 2);
```

```python
# New dataset
heart_disease = pd.read_csv("data/heart-disease.csv")
heart_disease.head()
```

```python
# Create a histogram of age
heart_disease["age"].plot.hist(bins=30)
```

```python
heart_disease.plot.hist(subplots=True, figsize=(10, 20));
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease["age"]> 50]
```

```python
over_50
```

```python
# Pyplod method
over_50.plot("age", "chol", kind="scatter", c="target")
```

```python
# OO method mixed with pyplot

fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
            ax=ax)
# ax.set_xlim([45, 100])
```

```python
# OO method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart disease and Cholesterol Levels in people over 50",
       xlabel="Age",
       ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(), linestyle="--")
```

```python
# Subplot of chol, age, thalach

fig, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(10, 12));


# Plot the data to ax1
scatter = ax1.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"])

# Customize ax1
ax1.set(title="Heart Disease and Cholesterol Levels in people over 50",
       xlabel="Age",
       ylabel="Cholesterol");

# add legend to the figure
fig.legend(*scatter.legend_elements(), title="Target");
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold")

# Add a horizontal line
ax1.axhline(over_50["chol"].mean(), linestyle="--");


# Plot the data to ax2
scatter = ax2.scatter(x=over_50["age"], y=over_50["thalach"], c=over_50["target"])
ax2.set(title="Heart Disease and Max Heart Rate in people over 50",
       xlabel="Age",
       ylabel="Max Heart Rate")
ax2.axhline(over_50["thalach"].mean(), linestyle="--");
```
